```python
import json

from bson.son import SON
from pymongo import MongoClient
from bson.code import Code
from datetime import datetime
```

```python
mongo_client = MongoClient("www16436ui.sakura.ne.jp:27017")
db=mongo_client['my_mongo']
```

```python
cols = db.collection_names()
for c in cols:
    print(c)
```

```python
for tweet in db.tweet.find():
    print(tweet)
```

```python
#count
print(db.tweet.find().count())
```

```python
#distinct
db.tweet.distinct('search')
```

```python
start = datetime(2016, 7, 17, 0,0,0,0)
end = datetime(2016, 7, 18, 0,0,0,0)

#where
for tweet in db.tweet.find({'created_at': {'$gt': start, '$lt': end}}, {'_id':0, 'text':1, 'created_at':1}):
    print(tweet)
```

```python
start = datetime(2016, 7, 17, 0,0,0,0)
end = datetime(2016, 7, 18, 0,0,0,0)

#グループごとの件数表示
pipeline = [
    {"$match": {"created_at": { "$gt": start ,  "$lt": end}}},
    {"$group": {"_id": "$search",
                "count": {"$sum": 1}}},
    {"$sort": SON([("count", -1), ("_id", -1)])}
]
list(db.tweet.aggregate(pipeline))
```

```python
start = datetime(2016, 7, 17, 0,0,0,0)
end = datetime(2016, 7, 18, 0,0,0,0)

#limit
for tweet in db.tweet.find({'created_at': {'$gt': start, '$lt': end}}, {'_id':0, 'text':1, 'created_at':1}).limit(10):
    print(tweet['created_at'])
```

```python
start = datetime(2016, 7, 17, 0,0,0,0)
end = datetime(2016, 7, 18, 0,0,0,0)

#skip limit
for tweet in db.tweet.find({'created_at': {'$gt': start, '$lt': end}}, {'_id':0, 'text':1, 'created_at':1}).skip(5).limit(10):
    print(tweet)
```
